```python
from xmlrpc.server import SimpleXMLRPCServer
from xmlrpc.server import SimpleXMLRPCRequestHandler
from xmlrpc import client as xc

import datetime

class RequestHandler(SimpleXMLRPCRequestHandler):
    rpc_paths = ('/RPC2',)

proxy = SimpleXMLRPCServer(("localhost", 8001), 
                            requestHandler=RequestHandler)

server = xc.ServerProxy("http://localhost:8008")
stats_server = xc.ServerProxy("http://localhost:8018")

proxy.register_introspection_functions()

# Добавление в лог через сервер
def add_log(log_line, start_time, end_time):
    try:
        stats_server.add_log(log_line,start_time, end_time)
    except:
        return False
    return True

def ping():
    event_start_time = datetime.datetime.now()
    res = server.ping()
    event_time = datetime.datetime.now() - event_start_time
    add_log('ping',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(ping, 'ping')

def now():
    event_start_time = datetime.datetime.now()
    res = server.now()
    event_time = datetime.datetime.now() - event_start_time
    add_log('now',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(now, 'now')

def show_type(arg):
    event_start_time = datetime.datetime.now()
    res = server.type(arg)
    event_time = datetime.datetime.now() - event_start_time
    add_log('type',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(show_type, 'type')

def test_sum(a, b):
    event_start_time = datetime.datetime.now()
    res = server.sum(a, b)
    event_time = datetime.datetime.now() - event_start_time
    add_log('sum', event_start_time, str(event_time))
    return res
proxy.register_function(test_sum, 'sum')

def test_pow(a, b):
    event_start_time = datetime.datetime.now()
    res = server.pow(a, b)
    event_time = datetime.datetime.now() - event_start_time
    add_log('pow',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(test_pow, 'pow')

def black_list_check(sname):
    event_start_time = datetime.datetime.now()
    res = server.black_list_check(sname)
    event_time = datetime.datetime.now() - event_start_time
    add_log('black_list_check',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(black_list_check, 'black_list_check')

def black_list_check_fio(sname, name, patronim, date):
    event_start_time = datetime.datetime.now()
    res = server.black_list_check_fio(sname, name, patronim, date)
    event_time = datetime.datetime.now() - event_start_time
    add_log('black_list_check_fio',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(black_list_check_fio, 'black_list_check_fio')

def send_back_binary(bin_data):
    event_start_time = datetime.datetime.now()
    res = server.send_back_binary(bin_data)
    event_time = datetime.datetime.now() - event_start_time
    add_log('send_back_binary',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(send_back_binary, 'send_back_binary')

def binarization(bin_data, threshold):
    event_start_time = datetime.datetime.now()
    res = server.binarization(bin_data, threshold)
    event_time = datetime.datetime.now() - event_start_time
    add_log('binarization',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(binarization, 'binarization')

def reflection_horizontal(bin_data):
    event_start_time = datetime.datetime.now()
    res = server.reflection_horizontal(bin_data)
    event_time = datetime.datetime.now() - event_start_time
    add_log('reflection_horizontal',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(reflection_horizontal, 'reflection_horizontal')

def reflection_vertical(bin_data):
    event_start_time = datetime.datetime.now()
    res = server.reflection_vertical(bin_data)
    event_time = datetime.datetime.now() - event_start_time
    add_log('reflection_vertical',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(reflection_vertical, 'reflection_vertical')

def send_back_inversion(bin_data):
    event_start_time = datetime.datetime.now()
    res = server.color_inversion(bin_data)
    event_time = datetime.datetime.now() - event_start_time
    add_log('color_inversion',event_start_time.strftime("%Y-%m-%d %H:%M:%S"), str(event_time))
    return res
proxy.register_function(send_back_inversion, 'color_inversion')
    
print("Listening on port 8001...")
proxy.serve_forever()
```
